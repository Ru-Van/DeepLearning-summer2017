## Scattor plots for 2D data


```octave
rand('twister',0) # rand('twister',0) is no longer used in Matlab and is replaced with the rng() function
X = 5 * rand(100,1); # your x data
y = rand * X + rand(100,1); % your y data
f1 = figure;       % create a blank figure
p1 = plot(X,y,'.') % plot x vs y
X1 = X;
```

```octave
# to make the graph looks more legit
#clf(f1) # clear the figure
p2 = plot(X,y,'o','MarkerEdgeColor','k','MarkerFaceColor','r','MarkerSize',5);
```

```octave
# to add title and axis label to the graph
% axis([leastX,maxX,leastY,maxY])
axis([-1,6,-1,5]);
xlabel('x'); 
ylabel('y');
title('this is the title');
```

## Plotting Curves:
- plotting curves or functions is very similar to the data plotting we just performed.
- We evaluate the function at every point along its domain and plot the resulting x,y pairs, connecting consecutive dots. 
- the ** hold on ** command in matlab tells Matlab to superimpose all the plots onto the same figure, rather than overwriting previous plots

```octave
f = @(x) x.^2 ;% this creates a function of x, namely f(x) = x .^2
g = @(x) sin(x) + 5;      % g(x) = 5*sin(x) +5
res = .001;  % resolution of the plot
domain = 0:res:2*pi;
f2 = figure; % open a new figure
p3 = plot(domain,f(domain),'Color','r');
hold on;
p4 = plot(domain,g(domain));
axis([0,2*pi,0,2*pi]);
xlabel('my x axis')
ylabel('my y axis')
title('Awesome Graph')
```

```octave
g = @(x) 5*sin(x) + 5;  
domain = -pi:res:pi;
f3 = figure('Color',[1,1,1]);
hold on;
p5 = plot(domain,f(domain),'--r','LineWidth',3); % plot a thick dashed red line
p6 = plot(domain,g(domain),'-b','LineWidth',3); % plot a thick solid blue line, 'b' stands for blue, and 'k'stands for black
subdomain = domain(1:400:end);
oneSTD = std(g(domain)); % standard deviation of g(domain)
oneSTD = oneSTD * ones(size(subdomain)) ;% make correct size, as needed by errorbar
p7 = errorbar(subdomain, g(subdomain),oneSTD);
title('LIT Graph','FontSize',12);
axis([-3,3,-5,15]);
grid on; % add grid lines
legend('f','g');
```

## Histograms




```octave
figure;
% Y = fix(X) rounds each element of X to the nearest integer toward zero.
% For positive X, the behavior of fix is the same as floor. 
% For negative X, the behavior of fix is the same as ceil.
grades = fix(normrnd(70,10,200,1)); % normrnd: normal random number, normrnd(mu, sigma,row, column)
hist(grades);
xlabel('percent'); ylabel('count');
title('grade distribution','fontsize',15);
```

## 3D plots
- plot3(X1,Y1,Z1): where X1, y1,Z1 are vectors or matrices

```octave
t = 0:pi/50:10*pi;
size(t)
st = sin(t);
ct = cos(t);
figure
plot3(st,ct,t,'r.')
figure
plot3(st,ct,t,'--c')
```

## [LineSpec](https://www.mathworks.com/help/matlab/ref/linespec.html)
- there are many instances where you could specify the linespec, both in 2d and 3d
- You can modify the three components:
    - line style
    - marker symbol
    - color
- Ex: plot(x,y,'d') -> this is plotting without a line style, so only the markers are plotted.
- ** Line Style Specifiers **
    1. '-':solid line (default)
    2. '--': dashed line
    3. ':'dotted line
    4. '-.'dashed dot line
- ** Markers specifiers **
    1. '+': plus sign
    2. 'o' circle
    3. '*' asterisk
    4. '.' point
- ** Color**:
    - 'r','g','b','c':cyan,'m':magenta,'y','k','w'
    
- ** Other related properties:**
    1. LineWidth: specify the width of the line
    2. MakerEdgeColor:color of the marker or the edge color for filled markers
    3. MarkerFaceColor: color of the face of filled marker
    4. MarkerSize: specify the size of the marker
