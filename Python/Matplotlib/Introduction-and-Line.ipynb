```python
x = [1,2,3]
y = [5,7,4]
x2 = [1,2,3]
y2 = [10,14,12]
plt.plot(x,y, label = "First Line")
plt.plot(x2,y2, label = "Second Line")
plt.xlabel('Plot Number')
plt.ylabel('Important var')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
plt.show()
```

    Bar charts and Histogram

```python
# This is bar chars, and notice that there is a gap in between
a = [2,4,6,8,10]
b = [6,7,8,2,4]
c = [1,3,5,7,9]
d = [7,8,2,4,2]
plt.bar(a,b, label = "Bars 1",color = 'r')
plt.bar(c,d,label = "Bar 2",color = 'c')
plt.show()
```

     Stack plots

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.stackplot(days, sleeping,eating,working,playing, colors =['m','c','r','k'])
plt.show()
```

    Pie Charts

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['b','r','c','m']
plt.pie(slices, labels = activities,colors = cols,shadow = True,explode = (0,0.2,0,0), autopct = '%1.1f%%')
plt.show()
```

```python
# you can use csv to open txt file 
x = []
y = []
with open('example.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        
plt.plot(x,y,label= 'loaded from file')
plt.show()
```

```python
#load the data using numpy
import numpy as np

x , y = np.loadtxt('example.txt',delimiter = ',', unpack = True)
plt.plot(x,y,label= 'loaded from file')
plt.show()
```
